# 4.Data Cleaning Operations

```python
def clean_data():
    while True:
        print("\n1.Check Missing value")
        print("2.Check Duplicated value")
        print("3.Check Descriptive Statistics")
        print("4.Drop Columns ")
        print("5.fill missing value with median(numeric columns)")
        print("6.Fill missing value with specific value")
        print("0.Exit\n")

        ch = int(input("Enter your choice"))

        if ch == 1:
            res=df.isnull().sum()
            print(res)
        elif ch == 2:
            res=df.duplicated().sum()
            print(res)
        elif ch == 3:
            res=df.describe()
            print(res)
        elif ch == 4:
            print(df.columns)
            col = input("Enter columns name:")
            df.drop(columns=[col],inplace=True)
            print(f" columns {col} Dropped Successfully....")
            print("=======columns=======")
            print(df.columns)
        elif ch == 5:
            print("\n===Fill missing value with median=====")
            col=input("Enter Columns name")
            df[col].fillna(df[col].median(),inplace=True)
            print("\n======Display missing value==========")
            res=df.isnull().sum()
            print(res)
        elif ch == 6:
            print("\n===Fill missing value with Specific value=====")
            col = input("Enter columns name:")
            value = input("Enter value:")
            df[col].fillna(value,inplace=True)
            print("\n======Display missing value==========")
            res=df.isnull().sum()
            print(res)
        elif ch == 0:
            break
clean_data()
```

# 5. Statistical Operations

```python
def Statistical_Operations():
    while True:
        print("1.Mean")
        print("2.Medain")
        print("3.Mode")
        print("4.Correlation")
        print("5.Standard Deviation")
        print("6.Variance")
        print("7.Exit")
    
        ch = int(input("Enter your choice"))
    
        if ch == 1:
            res = df.mean(numeric_only=True)
            print(res)
        elif ch == 2:
            res = df.median(numeric_only=True)
            print(res)
        elif ch == 3:
            col = input("Enter columns name(Categorical columns):")
            res = df[col].mode()
            print(res)
        elif ch == 4:
            res = df.corr(numeric_only=True)
            print(res)
        elif ch == 5:
            res = df.std(numeric_only=True)
            print(res)
        elif ch == 6:
            res= df.var(numeric_only=True)
            print(res)
    
        elif ch == 7:
            break
Statistical_Operations()
        
    
```

# 6. Grouping and Aggregation Operations

```python
def grouping_operation():
    while True:
        print("\n1.Region-wise Average Happiness Score")
        print("2.Region-wise Multiple Statistics (Mean, Max, Min)(pivot table)")
        print("3.Region-wise GDP Comparison (Bar Plot)")
        print("4.Region vs Country – Happiness Score")
        print("5.Region-wise GDP vs Happiness")
        print("6.Exit\n")
    
        ch = int(input("Enter your choice:"))

        if ch == 1:
            region_avg = df.groupby("Region")["Happiness Score"].mean()
            print(region_avg)
        elif ch == 2:
            res = pd.pivot_table(df,values="Happiness Score",index="Region",aggfunc=["mean","max","min"])
            print(res)
        elif ch == 3:
            res=df.groupby("Region")["Economy (GDP per Capita)"].mean()
            print(res)
        elif ch == 4:
            res=pd.pivot_table(df,values="Happiness Score",index="Region",columns="Country",aggfunc="mean")
            print(res)
        elif ch == 5:
            res = pd.pivot_table(df,values=["Happiness Score","Economy (GDP per Capita)"],index="Region",aggfunc="mean")
            print(res)
            
        elif ch == 6:
            break
grouping_operation()
        
```

# 7. Data Selection & Column Operations

```python
def data_selection():
    while True:
        print("\n1.Columns Selection")
        print("2.Row Selection (slicing)")
        print("3.Filter Data")
        print("4.Rename Columns")
        print("5.Exit\n")
    
        ch = int(input("Enter your choice:"))
        if ch == 1:
            col = input("Enter column name:")
            print(df[col])
        elif ch == 2:
            s = int(input("Enter start index:"))
            e = int(input("Enter end index:"))
            print(df.iloc[s:e])
        elif ch == 3:
            value = float(input("Enter value (Happiness Score > value):"))
            print(df[df["Happiness Score"] > value])
        elif ch == 4:
            print("Columna names :",df.columns)
            old = input("Enter old column name:")
            if old not in df.columns:
                print("Column name not found")
                return
            new = input("Enter new column name:")
            df.rename(columns={old:new},inplace=True)
            print("\nColumn name changed successfully\n")
            print("Columna names :",df.columns)
            
        elif ch == 5:
            break
data_selection()
    
```

# 8. Data Visualization

```python
def data_visuliazation():
    sns.set_theme(style="whitegrid",context="paper",font="cursive",font_scale=1.2)
    while True:
        print("1.Bar plot")
        print("2.Pie Plot")
        print("3.line plot")
        print("4.scatter plot")
        print("5.Histogram plot")
        print("6.Heat Map")
        print("7.Box Plot")
        print("8.Subplot")
        print("9.go back")

        ch = int(input("Enter your choice:"))

        if ch == 1:
            sns.barplot(y="Region",x="Economy (GDP per Capita)",errorbar=('ci',True),data=df)
            plt.show()
        elif ch == 2:
            region_avg = df.groupby("Region")["Happiness Score"].mean()
            plt.pie(region_avg,labels=region_avg.index,autopct="%1.2f%%")
            plt.legend(loc="upper right",fontsize=5)
            plt.show()
        elif ch == 3:
            plt.plot(df["Happiness Rank"],df["Happiness Score"])
            plt.show()
        elif ch == 4:
            sns.scatterplot(x="Economy (GDP per Capita)",y="Happiness Score",hue="Region",data=df)
            plt.legend(fontsize=5)
            plt.show()
        elif ch == 5:
            a=sns.histplot(x="Happiness Score",data=df)
            for i in a.containers:
                a.bar_label(i)
            plt.show()
        elif ch == 6:
            sns.heatmap(df.corr(numeric_only=True),annot=True)
            plt.show()
        elif ch == 7:
            sns.boxplot(y="Happiness Score",data=df)
            plt.show()
        elif ch == 8:
            
            fig,a = plt.subplots(2,3,figsize=(24,22),dpi=200)

            
            p=sns.barplot(y="Region",x="Economy (GDP per Capita)",data=df,ax=a[0,0],errorbar=('ci',True))
            for i in p.containers:
                p.bar_label(i,fmt="%1.1f%%")
            a[0,0].set_title("Bar Plot")

            region_avg = df.groupby("Region")["Happiness Score"].mean()
            a[0,1].pie(region_avg,labels=region_avg.index,autopct="%1.2f%%")
            a[0,1].legend(loc="upper right",fontsize=5)
            a[0,1].set_title("Pie Plot")
            
            a[0,2].plot(df["Happiness Rank"],df["Happiness Score"])
            a[0,2].set_title("Line plot")

            sns.scatterplot(x="Economy (GDP per Capita)",y="Happiness Score",hue="Region",data=df,ax=a[1,0])
            a[1,0].legend(fontsize=5)
            a[1,0].set_title("Scatter Plot")
          

            p=sns.histplot(x="Happiness Score",data=df,ax=a[1,1])
            for i in p.containers:
                p.bar_label(i)
            a[1,1].set_title("Histogram")

            sns.boxplot(y="Happiness Score",data=df,ax=a[1,2])
            a[1,2].set_title("Boxplot")

            plt.tight_layout()
            plt.show()
            
            
        elif ch == 9:
            break
data_visuliazation()  
            
```
